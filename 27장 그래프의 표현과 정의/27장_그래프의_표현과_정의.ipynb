### 그래프의 정의
그래프 G(V,E)  
- 집합 V : 어떤 자료나 개념을 표현하는 정점(vertex)들의 집합

-> 그래프는 정점들과 이들을 연결하는 간선으로 구성됨  
-> 정점의 위치정보, 간선순서는 그래프 정의에 포함 X

### 그래프의 종류
1) 방향 그래프, 유향 그래프
- **방향 그래프**  
각 간선 : 방향이라는 새로운 속성  
두 정점 u,v 가 있을 때 u->v 와 v->u 서로 다른 간선
ex. 짝사랑, 일방통행
- **무향 그래프**  
간선에 방향이 없는 그래프
ex. 트리 , 루트 없는 트리 (간선 연결 관계만 보면)

2) 가중치 그래프
- **가중치 그래프**
각 간선에 가중치(실수 속성) 부여
ex. 두 사람 사이의 호감도, 두 도시 사이의 거리

3) 그래프의 형태로 그래프를 분류하는 경우
- **다중 그래프**
두 정점 사이에 두 개 이상의 간선이 있을 수 있는 그래프
ex. 두 지점을 잇는 두 개 이상의 도로
- **단순 그래프**
두 정점 사이에 최대 한 개의 간선만 있는 그래프

4) 이분 그래프
- **이분 그래프**  
그래프의 정점들을 겹치지 않는 두 개의 그룹으로 나눠서 서로 다른 그룹에 속한 정점들 간에만 간선이 존재하도록 만들 수 있는 그래프

5) 여러 가지 속성을 함께 갖는 그래프    
ex. 방향 가중치 그래프, 이분 가중치 그래프

 * **사이클 없는 방향 그래프 (DAG)**  
 > 방향 그래프, 한 점에서 출발해 자기 자신으로 돌아오는 경로가 존재하지 않는 경우  
 ex. 여러 직업들 간 상호 의존 관계 표현

##27.2 그래프의 사용 예    

**철도망의 안정성 분석**  
여러 객체 사이의 연결관계  
철도 각 역을 정점으로, 두 역 사이를 연결하는 철로를 간선으로 표현하는 그래프 만들고 절단점 찾기 알고리즘 이용하여 해결  


**소셜 네트워크 분석**   
사람들 간 지인/친밀도 관계 분석  
SNS 회원들을 정점으로, 두 회원이 서로 친구 사이인 경우 간선으로 연결하는 그래프 만들기  
그래프의 너비 우선 탐색 이용하여 내가 한 다리 건너면 알게 되는 사람, 몇 다리를 건너야 누군가와 아는 사이인지 알아볼 수 있음  

**인터넷 전송속도 계산**  
컴퓨터들과 라우터들이 네트워크로 연결된 것 그래프로 표현 가능  
인터넷에 연결된 라우터와 컴퓨터들을 정점으로, 두 기계 연결하는 케이블을 간선으로 표현하는 그래프 만들기 가능  

**한 붓 그리기**  
선들이 만나는 점을 정점으로, 이 점들을 연결하는 선들을 간선으로 삼은 그래프 만들기  
오일러 경로라 부르는 문제, 깊이 우선 탐색 응용하여 해결 가능  

**외환 거래**  
각 통화를 정점으로, 서로 교환 가능한 통화 사이에 간선을 두는 방향 그래프 만들기  
U->V로 가는 간선의 가중치는 두 통화 간 교환 비율 나타냄



##27.3 암시적 그래프 구조들  
- 현실 세계에서 그래프 같은 형태를 갖는 구조가 아니라도 그래프를 통해 표현하면 쉽게 해결 가능한 문제들 존재 : 암시적 그래프  

- 예시  

> **할 일 목록 정리**  
할 일 어떤 순서로 해야 하는지 문제 : 위상 정렬 문제, 깊이 우선 탐색 응용해 풀이 가능  

**15-퍼즐**  
4*4 게임판에 1~15까지의 숫자가 적힌 열다섯 개의 타일 끼워져 있음, 이 타일들 움직여 원래 있던 순서대로 정렬하는 문제  
게임판에서 현재 타일의 배치를 하나의 정점으로, 한 배치에서 타일을 한 번 움직여 다른 배치 얻기 가능할 때 두 정점을 간선으로 연결하는 그래프 상상하기  
그래프 상에서 두 점 사이 잇는 가장 짧은 경로 구하는 문제  

**게임판 덮기**  
NxN 정사각형의 게임판을 1X2 크기의 블록으로 덮는 문제  
막히지 않은 게임판의 각 칸을 정점으로, 상하좌우로 인접한 칸들 사이에 간선 연결하는 그래프 만들기 : 이분 그래프 문제, 이분 매칭 알고리즘으로 풀이  


**회의실 배정**  
N개 팀이 회의할 때 회의실이 하나이고 각 팀이 사용하고자 하는 시간 두 개씩 적어 낸다  
각 팀마다 적어낸 두 시간 중 하나씩 배정해서 모든 팀이 회의하도록 하는 문제 : 만족성 문제  
두 선택지 중 하나를 모든 사람이 선택해야 하는 문제  : *2-SAT*  
: 그래프에서 강 결합성 문제로 변환하여 풀이 가능


##27.4 그래프의 표현 방법

그래프는 트리에 비해 *정적인 용도*로 사용됨
-> 새로운 정점이나 간선을 추가하고 삭제하는 일이 자주 일어나지 않음  
-> 그래프는 구조의 변경이 어렵더라도 *간단한 방법*으로 구현  

**간단한 방법**  
그래프 정점을 객체의 인스턴스로 표현  
각 정점에 0부터 시작하는 번호 붙이기  
배열에 각 정점 정보 저장  
각 간선은 반대쪽 끝 정점의 번호 저장  


그래프의 두 표현 방식 : *인접 리스트 표현, 인접 행렬 표현*

###인접 행렬 표현 vs 인접 리스트 표현  
1) 인접 행렬 표현의 장점  

번호 u,v 가 주어졌을 때 두 장점 잇는 간선 있는지 여부를 한 번의 배열 접근만으로 확인 가능  


2) 인접 리스트 표현 단점  

인접 리스트는 간선 (u,v) 존재 여부 확인 위해 연결리스트 adjacent[u] 처음부터 각 원소 일일이 확인 필요  


3) 인접 행렬 표현 단점  

|V| X |V|의 2차원 배열 사용하기 때문에 실제 간선 개수와 관계없이 항상 O(|v|^2) 크기의 공간 사용함  

4) 인접 리스트 표현 장점  

인접 리스트 표현은 |V| 개의 연결 리스트에 실제 간선 수만큼 원소가 들어 있어 전체 **O(|V| + |E|)** 의 공간만 사용   

간선의 수가 |V|^2과 비슷하면 비슷한 양의 메모리 쓰지만 간선의 수가 훨씬 적으면 결과적으로 훨씬 효율적  

**희소 그래프** : 간선의 수가 |V|^2보다 훨씬 적은 그래프  
**밀집 그래프** : 간선의 수가 |V|^2에 비례하는 그래프  

희소 그래프에는 인접 리스트, 밀집 그래프에는 인접 행렬 사용하는 것이 유리  








###암시적 그래프 표현


그래프로 푸는 문제라고 항상 그래프를 메모리에 직접 표현해야 할 필요 없음  

암시적 그래프 : 그래프 구조 쓰지 않고 문제 해결  

ex. 그리드 문제를 그래프로 처리 -> 격자판을 그래프로 나타내면 각 격자를 노드로, 인접한 격자를 간선으로 나타내기  


비교적 단순한 문제에서 입력 일일이 그래프로 표현하기 번거롭거나  그래프 크기가 아주 큰데 실제로는 그 중 일부만 사용하는 경우 암시적 그래프 표현 사용이 유리  
암시적 그래프 표현은 코드가 복잡해진다는 문제점
